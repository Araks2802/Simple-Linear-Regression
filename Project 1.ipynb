```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sklearn as sk
```

```python
data = pd.read_csv("C:\\Users\\Vandana\\Downloads\\Student_Performance.csv")
data
```

```python
data = data[['Hours Studied','Previous Scores',"Sleep Hours","Sample Question Papers Practiced","Performance Index"]]
data
```

```python
x = np.array(data.drop(['Performance Index'],axis = 'columns'))
y = np.array(data['Performance Index'])

x_train, x_test, y_train, y_test = sk.model_selection.train_test_split(x,y,test_size = 0.1)

linear = sk.linear_model.LinearRegression()
linear.fit(x_train,y_train)
acc = linear.score(x_test,y_test)
print(acc)

print(linear.coef_)
print(linear.intercept_)
```

```python
predictions = linear.predict(x_test)

for i in range(0,20):
    print(predictions[i],x_test[i],y_test[i])
```

```python
d = []
HS = input("Enter number of hours studied: ")
PS = input("Enter previous score: ")
SH = input("Enter amount of hours slept: ")
SP = input("Enter number of sample papers practised: ")
d.append(HS)
d.append(PS)
d.append(SH)
d.append(SP)

p = linear.predict(np.array(d).reshape(1,-1))
if (p<=0):
    print("\n\nPerformance Index: 0")
elif (p>=100):
    print("\n\nPerformance Index: 100")
else:
    print("\n\nPerformance Index: ",p[0])
```
